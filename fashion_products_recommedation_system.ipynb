```python
# Importing Required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("fashion_products.csv")
test_data = pd.read_csv("fashion_products.csv")
```

```python
data.head()
```

```python
#Info About dataset
data.info()
```

```python
#Finding null values
data.isnull().mean()*100
```

```python
#Finding Duplicates
data.duplicated().sum()
```

```python
#Finding categorical columns
cat_col=[]
for col in data.columns:
    if(data[col].dtypes=='object'):
        cat_col.append(col)
```

```python
#Finding numerical columns:
num_col=[]
for col in data.columns:
    if(data[col].dtypes!='object'):
        num_col.append(col)
```

```python
#Reset index
data.set_index('Product ID',inplace=True)
```

```python
#Droping Unnecessary Features
data.drop(columns=['User ID'],inplace=True)
```

```python
#Removing columns that are not required further in num_col
num_col.remove('Product ID')
num_col.remove('User ID')
```

```python
#Encoding Categorical Columns
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
cat_class=[]
for col in cat_col:
    data[col]=encoder.fit_transform(data[col])
    cat_class.append(encoder.classes_)
```

```python
#Standardization of numerical columns
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
for col in num_col:
    data[col]=scaler.fit_transform(np.array(data[col]).reshape(len(data[col]),1))
```

```python
#After all preprocessing work required for recommedation system
data.head()
```

```python
#Model Creation for recommedation system
from sklearn.metrics.pairwise import cosine_similarity
similar=cosine_similarity(data)
```

```python
#Function for input output in recommedation system
def recommend(product_id):
    index=product_id-1
    item=sorted(list(enumerate(similar[index])),key=lambda x:x[1],reverse=True)[1:6]
    id=[]
    for i in item:
        id.append(i[0])
    return test_data.iloc[id]
```

```python
#Input
product_id=int(input("Enter Product ID"))
result=recommend(product_id)
```
